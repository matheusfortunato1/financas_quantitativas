# Juros

Preço do dinheiro no tempo. A taxa de juros basica é o fator principal na precificação de qualquer ativo ou derivativo. 

Taxa de juros depende do risco de crédito, que significa o risco de inadimplencia. Quanto maior o risco, maior a taxa.

A taxa de juros livre de risco é a que não há risco de crédito, normalmente titulos publicos emitidos pelo governo (risco de default, apenas).

Suponha que uma quantia A é investida por n anos ha uma taxa de R ao ano. Então o valor final do investimento é dado por

$$ A(1+R)^n \tag{1}$$

O juros acima é conhecido a juros compostos de capitalização discreta. Existe também o de capitalização continua, onde o valor final é dado por:

$$ Ae^{Rn} \tag{2}$$

## Taxa de cupom-zero
Taxa de juros em que não há pagamento de cupons intermediarios. Ha apenas o pagamento do valor final do titulo. As vezes, é referenciada como *taxa spot*. 

Os titulos de zero cupom emitidos pelo governo determinado as taxas à vista para um numero finito de maturidades.

## Taxa Forward (a termo)
É a taxa prevista hoje para vigorar entre duas datas futuras.


## Valor presente liquido 
Valor de todos os pagamentos realizados do instante inicial até o de maturação do contrato com valor descontado para o ponto inicial a uma taxa de juros constante. A taxa de juros é chamada de taxa de desconto do fluxo de caixa.

Para capitalização discreta: 
$$ PV = \sum^T_{t=0} \frac{FC_{t}}{(1+r_{d})^t} \tag{3}$$

Para capitalização continua
$$ PV = \sum^T_{t=0} FC_{t}e^{-r_{c}t} \tag{4}$$

## Precificação de titulos
A maioria de titulos pagam cupons periodicamente. O proprietario do titulo recebe o valor principal e também o cupom na sua data de maturidade(?). O preço teorico do titulo pode ser calculado como *present value* de todo fluxo de caixa que será recebido pelo dono, usando uma taxa de desconto apropriada. 

$$ P_{0} = \sum^T_{t=1} \frac{C_{t}}{(1+r_{t})^t} \tag{5}$$

Perceba que a taxa $r_{t}$ varia.

### Bond yield (TIR - Taxa interna de retorno)
É a taxa de desconto que iguala o fluxo de caixa descontado ao valor de mercado do titulo.

$$ - P_{0} + \sum^T_{t=1} \frac{C_{t}}{(1+y)^t} = 0 \tag{6}$$

Perceba que a taxa $y$ é constante.

### Cupom yield

$$ \alpha = \frac{C}{F}\% \tag{7}$$

- Se ($ TIR > \alpha $) -> titulo está sendo negociado com desconto ou deságio
- Se ($ TIR = \alpha $) -> titulo está sendo negociado ao par
- Se ($ TIR < \alpha $) -> titulo está sendo negociado com prêmio ou ágio

### Par yield

É a taxa de cupom que faz com que o present value do titulo seja igual ao valor de face. 

$$ \sum^{(T-1)}_{t=1} \frac{C}{(1+r_{t})^t} + \frac{C+F}{(1+r_{T})^T} = F \tag{8}$$

$$ Par yield = \frac{C}{F}\% \tag{9}$$

- Se ($ RP > \alpha $) -> titulo está sendo negociado com desconto ou deságio
- Se ($ RP = \alpha $) -> titulo está sendo negociado ao par
- Se ($ RP < \alpha $) -> titulo está sendo negociado com prêmio ou ágio

# Sensibilidade de preços

A análise de sensibilidade permite avaliar o impacto na variação do preço do titulo causado por uma mudança na estrutura da taxa de juros. Por simplificação, **a analise é realizada atraves da variação da yield (TIR)**

Varia-se na equação *(6)* a taxa $y$ e captura-se a variação no preço $P_{0}$

Outra maneira de se calcular a sensibilidade é através da *duration*.

## Duration

É uma medida de quanto tempo em média o dono do titulo precisa esperar para receber o pagamento. De maneira mais formal, é o tempo ponderado pelo fluxo de caixa em relação ao preço do titulo. Ou seja:

$$ D = \frac{1}{P}{\sum^T_{t=1} t*\frac{FC_{t}}{(1+y)^t} } \tag{10}$$
ou
$$ D = \frac{\sum^T_{t=1} t*FC_{t}*e^{-yt_{i}}}{P} \tag{11}$$
Então
$$ \frac{dP}{dy} = -\sum^n_{i=1} c_{i}t_{i}e^{-yt_{i}} \tag{12} $$
$$ \frac{dP}{dy} = -PD \tag{13} $$

**Duration modified** é definida por:
$$ D^* = -\frac{1}{P}\frac{dP}{dy} \tag{14} $$

Utilizando o metodo de aproximação de Taylor e expandindo a equação *(10)* temos
$$ \Delta P = -P*D^**\Delta y + \frac{1}{2}P*C^**(\Delta y)^2 \tag{15} $$

Onde C* é a Convexidade.

Fazendo a leitura da equação *(13)* percebe-se que faz total sentido. Se a taxa de juros sobe, o valor do titulo cai. 

## Estrutura a termo da taxa de juros (ETTJ)

Para n datas distintas de maturidade, as taxas a vista podem ser estimadas através de uma interpolação. Essa curva interpolada é conhecida como *curva zero* ou ETTJ.

Os titulos zero cupom são conhecidos como vértices da ETTJ.

Representa as expectativas do custo do dinheiro para diferentes prazos de vencimento. Existe uma ETTJ para cada classe de titulo de um determinado emissor. Que pode ser aproximada adicionando um spread à curva zero.

Alguns comentários gerais:

- O governo interfere no formato da ETTJ através da emissão ou recompra dos seus titulos de renda fixa
- A ETTJ carrega a expectativa futura dos agentes economicos (governos, consumidores e firmas)
- Uma curva positivamente inclinada indica uma *economia normal* (contango)
- Uma curva horizontal indica nervosismo do mercado (flat)
- Uma curva negativamente inclinada indica um cenário pessimista ou inicio de uma crise (backwardation)

### Lei do preço único
Define a relação entre as taxas à vista e a termo. Afirma que dois titulos com o mesmo fluxo de caixa devem ter o mesmo preço. 

Seja $y_{n}$ a taxa a vista de 0 a n. E $f_{n}$ a taxa forward de n-1 a n. Então

$$ [1+y_{n}] = [1+y_{n-1}]^{n-1}[1+f_{n}] \tag{16} $$
ou
$$e^{y_{n}*n} = e^{y_{n-1}*(n-1)}*e^{f_{n}} \tag{17} $$

Somos levado a uma relação entre a taxa a vista $y_{n}$ é a média aritmética das taxas a termo:
$$y(n) = \frac{1}{n}\sum^n_{n=1}f(i) \tag{18} $$



As tentativas de explicar o formato da curva zero são várias e essas são algumas teorias:

### Teoria das expectativas do mercado 
As taxas de juros de longo prazo devem refletir as taxas de juros de curto prazo (como tudo na vida, o longo prazo é uma sucessão de curtos prazos).

### Teoria da segmentação de mercado
As taxas de curto, medio e longo prazo são determinadas de forma independente. Agentes do mercado tem preferencias por um determinado segmento da ETTJ

### Teoria da preferencia por liquidez
Os investidores preferem preservar sua liquidez e investir por breves periodos de tempo.
Os devedores preferem tomar empréstimos por taxas fixas e por longos periodos de tempo. 

## Modelagem da ETTJ

### Paramétricos
São interessantes para avaliar as movimentações e extrapolações da ETTJ
### Não-paramétricos
Interessante para avaliar a taxa a uma determinada maturidade. 

### Hedging portfolios of assets and liabilities

Instituições financeiras normalmente fazem o *hedge* do risco da taxa de juros através da duration. Isso é realizado garantindo que a duration média dos seus assets se iguala a duration média das suas liabilities. Ou seja, a média duration dos titulos em que a instituição é dona é igual a média da duration dos titulos short ou emitidos pela instituição.
